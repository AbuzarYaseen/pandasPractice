```python
import numpy as np 
import pandas as pd 
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
#Checking null values in columns
```

```python
df.isna().any()
```

```python
df
```

```python
# Removing missing data works on the basis of rows
```

```python
df.dropna()
```

```python
df
```

```python
df.dropna(thresh=3) #thresh=3 means 3 none null values in each row
```

```python
# Filling the missing data
```

```python
df
```

```python
df.fillna(0)
```

```python
values = {'A':10, 'B':20, 'C':30, 'D':40} # Filling is done in column vise
df.fillna(value=values)
```

```python
df
```

```python
df.fillna(df.mean())
```
