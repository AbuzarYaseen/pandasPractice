```python
#A series is a one dimentional labled array capable of holding any data type. The axis lables are collectively called index
#Series are vertical and arrays are horizontal.
```

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a', 'c', 'c']
myList = [10, 20, 30]
myArr = np.array([10, 20, 30])
dic = {1:10, 2:20, 3:30}
```

```python
pd.Series(myList) #Series can be created from list
```

```python
pd.Series(myList, index=labels)
```

```python
pd.Series(myArr, index=labels) #Series can alse be created from arrays
```

```python
pd.Series(dic) #Series can alse be created from dictionaries, creating from dictionary indexs are also setted from the keys of dictionary
```
