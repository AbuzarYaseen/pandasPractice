```python
import numpy as np
import pandas as pd 
```

# Merging of 2 Dataframes

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees, salaries,on='employee_id',how='inner')   #inner means merge tables having common employee_id
```

```python
pd.merge(employees, salaries,on='employee_id',how='outer')   #outer means combine the tables on the basis on employee_id
```

```python
#left means merge the tables on the basis of left table(employees)
pd.merge(employees, salaries,on='employee_id',how='left')
```

```python
#right means merge the tables on the right of right table(salaries)
pd.merge(employees, salaries,on='employee_id',how='right') 
```

# Concatination of 2 Dataframes

```python
df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

**Concatinating of columns**

```python
pd.concat([df1,df2])
```

**Concatinating of rows**

```python
pd.concat([df1, df2],axis=1)
```
