```python
# When multiple series are combined then this becomes a data frame.
# Data frame can be created from a dictionary and a list.
```

```python
import numpy as np
import pandas as pd
```

# Creating Data Frame

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(data)
df
```

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df2 = pd.DataFrame(data_list)
columns = ["Name","Age","City","Salary"]
df2 = pd.DataFrame(data_list,columns =columns)
df2
```

# Selection and Indexing of Columns

```python
df2["Name"]
```

```python
df2[["City", "Name"]]
```

**Creating a new column**

```python
df2["Designation"] = ["Engineer", "Doctor", "Plumber", "Pilot"]
```

```python
df2
```

**Removing a column**

```python
# Below is the snapshoot, column is still in the main data
df2.drop("Designation", axis=1)
```

```python
df2
```

**For deleting completly**

```python
df2.drop("Designation",axis=1,inplace=True)
```

```python
df2
```

```python
df2.drop(["Salary", "City"],axis=1)
```

```python
df2
```

```python
# Removing Rows same as removing columns just gave axis 0
df2.drop(0,axis=0)
```

```python
df2.drop([1,2],axis=0)
```

**Conditional Selection**

```python
# Only wants to see those people whose age is above 30
```
