```python
def extractEpisodes(txt):
    check = False
    data = ""
    for i in txt:
        if i == ")":
            check = False
            return data
        if check == True:
            data = data + i
        if i == "(":
            check = True
        
       
```

```python
df["Episodes"] = df['Title'].apply(extractEpisodes)  #apply runs on all the rows of data
```
